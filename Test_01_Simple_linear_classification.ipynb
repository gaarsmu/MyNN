```python
import numpy as np
import MyNN
import matplotlib.pyplot as plt
```

```python
X = (np.random.rand(2, 1000)-0.5)
Y = np.zeros((1,1000))
for i in range(1000):
    if X[0, i] >= X[1,i]:
        Y[0,i] = 1
```

```python
nn = MyNN.MyNN(2)
nn.add(1, 'Sigmoid')
nn.compile('Cross entropy')
```

```python
print(nn.f_script)
print(nn.b_script)
```

```python
nn.optimize(X,Y,lr=1, num_iterations=5000,report_cost=True)
```

```python
X = (np.random.rand(2, 1000)-0.5)
Y = np.zeros((1,1000))
for i in range(1000):
    if X[0, i] * X[1,i] >= 0:
        Y[0,i] = 1
```

```python
plt.scatter(X[0,:], X[1,:], c=Y, cmap=plt.cm.Spectral)
plt.show()
```

```python
nn = MyNN.MyNN(2)
nn.add(4, 'ReLU')
nn.add(1, 'Sigmoid')
nn.compile('Cross entropy')
```

```python
print(nn.f_script)
print(nn.b_script)
```

```python
print(nn.body)
```

```python
nn.optimize(X,Y,lr=0.1, num_iterations=10000,report_cost=True, report_cost_freq=500)
```
